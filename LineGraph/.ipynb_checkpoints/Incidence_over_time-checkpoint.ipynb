```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

#data on coronavirus from wikipedia (starting on the 10th of January)
cases_cumulative = [41, 41, 41, 41, 41, 45, 62, 121, 198, 291, 440, 571, 830, 1287, 1975, 2744, 4515, 5974, 7711, 9692, 11791, 14380, 17205, 20440, 24324, 28018, 31161, 34546, 37198, 40171, 42638]
deaths_cumulative = [1, 1, 1, 1, 2, 2, 2, 2, 3, 6, 9, 17, 25, 41, 56, 80, 106, 132, 170, 213, 259, 304, 361, 425, 490, 563, 637, 722, 811, 908, 1016]
```

So wikipedia, and many media outlets, have been providing cumulative values of incident cases and deaths relating to the recent outbreak of coronavirus. We can plot this...

```python
x_arr = np.arange(0, len(cases_cumulative), 1)#generate an integer for each day

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x_arr, cases_cumulative, '-k') #incidence in black (left axis)
ax2 = ax1.twinx()
ax2.plot(x_arr, deaths_cumulative, '-r')#deaths in red (right axis)
plt.show()
```

And then add some axis labels and a legend to make it clear what is going on...

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
h1, = ax1.plot(x_arr, cases_cumulative, '-k') #incidence in black (left axis)
ax2 = ax1.twinx()
h2, = ax2.plot(x_arr, deaths_cumulative, '-r')#deaths in red (right axis)

#axis labels
ax1.set_ylabel("Incident Cases (cumulative)")
ax2.set_ylabel("Deaths (cumulative)")
ax2.tick_params(axis='y', colors='red')#change axis tick colour to match plotted line
ax1.set_xlabel("Days")

#legend
plt.legend((h1, h2), ("Incidence", "Mortality"))

plt.show()
```

In order to understand how the epidemic is changing over time, however, I'm most interested in the number of new cases each day. 

Here is a function that converts cumulative cases per day to new cases per day. 

```python

    
def cumulative_to_new(cumulative_list):
    temp = cumulative_list.copy()
    temp.insert(0, 0) #zero in first day
    temp.pop() # remove last day
    new_list = np.subtract(cumulative_list,  temp)
    return new_list    

cases_cumulative = [41, 41, 41, 41, 41, 45, 62, 121, 198, 291, 440, 571, 830, 1287, 1975, 2744, 4515, 5974, 7711, 9692, 11791, 14380, 17205, 20440, 24324, 28018, 31161, 34546, 37198, 40171, 42638]
cases_new = cumulative_to_new(cases_cumulative)
    
deaths_cumulative = [1, 1, 1, 1, 2, 2, 2, 2, 3, 6, 9, 17, 25, 41, 56, 80, 106, 132, 170, 213, 259, 304, 361, 425, 490, 563, 637, 722, 811, 908, 1016]
deaths_new = cumulative_to_new(deaths_cumulative)

#set dates
date1 = mdates.datetime.date(2020, 1, 10)#date we started collecting data
date2 = mdates.datetime.date(2020, 2, 10)#yesterday's date
delta = mdates.datetime.timedelta(days=1)
dates = mdates.drange(date1, date2, delta)

    
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot_date(dates, cases_new, '-k')
ax2 = ax1.twinx()
ax2.plot_date(dates, deaths_new, '-r')
plt.show()
```

First thing, obviously, we need to make the ticklabels on the x-axis visible. 

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot_date(dates, cases_new, '-k')
ax2 = ax1.twinx()
ax2.plot_date(dates, deaths_new, '-r')
ax1.xaxis.set_tick_params(rotation=30)
plt.show()
```

And I would like them with a UK date format...

```python
formatter = mdates.DateFormatter('%d/%m/%y')#uk style date

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot_date(dates, cases_new, '-k')
ax2 = ax1.twinx()
ax2.plot_date(dates, deaths_new, '-r')
ax1.xaxis.set_major_formatter(formatter)#sets the format to the correct date type
ax1.xaxis.set_tick_params(rotation=30)
plt.show()
```

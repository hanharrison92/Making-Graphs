```python
import numpy as np
import matplotlib.pyplot as plt
cmap = plt.get_cmap('viridis')
cmaplist = [cmap(i) for i in range(cmap.N)]
```

```python
p2_base = 1-0.069-0-0.094-0.123-0.236-0.081-0.269
```

```python
p1 = [0.691, 0.727, 0.746]
p2 = [p2_base, 0.159, 0.691]
p3 = [0.691, 0.364, 0.159]
p4 = [0.814, 1, 1]
time=[0, 3, 12]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(time, p1, "--o", color=cmaplist[0], label="p1")
ax1.plot(time, p2, "--o", color=cmaplist[70], label="p2")
ax1.plot(time, p3, "--o", color=cmaplist[140], label="p3")
ax1.plot(time, p4, "--o", color=cmaplist[210], label="p4")
ax1.set_ylim([-0.02, 1.02])
ax1.set_xlim([-0.24, 12.24])
ax1.set_xlabel("Months")
ax1.set_ylabel("Utility Score (EQ-5D)")
ax1.legend(bbox_to_anchor=(1.01, 0.35), loc='lower left')
```

```python
def QALYs_calc(p_temp):
    QALY_3mnth = 0.5 * 0.25*(p_temp[0]+p_temp[1]) #three month calc
    QALY_1yr = 0.5 * 0.75*(p_temp[1]+p_temp[2])
    QALY_total = QALY_3mnth+QALY_1yr
    return [QALY_3mnth, QALY_1yr, QALY_total]
```

```python
print("\t\t0-3 months\t3-12 months\tTotal")
for indx, p in enumerate([p1, p2, p3, p4]):
    temp = QALYs_calc(p)
    #print(p)
    print("Patient %d\t%.4f\t\t%.4f\t\t%.4f" % (indx+1, temp[0], temp[1], temp[2]))
    
```
